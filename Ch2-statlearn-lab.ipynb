## Basic Commands

`Python` uses *functions* to perform operations. To run a function called `funcname`, we type `funcname(input1, input2)`, where the inputs (or *arguments*) `input1` and `input2` tell `Python` how to run the function. A function can have any number of inputs. Some functions are defined in modules, which is code made and packaged by someone else. The `numpy` module defines many useful functions for fast array operations. To use a module we must import it. For example, to create a vector (array) of numbers, we use the function `array()` from `numpy`. Any list of numbers inside the parentheses are transformed into and array. The following command
instructs `numpy` to create an array with the numbers 1, 3, 2, and 5, and to save them as a vector named `x`. When we type `x`, it gives us back the vector.

```python
import numpy as np

x = np.array([1, 3, 2, 5])
x
```

```python
x = np.array([1, 6, 2])
y = np.array([1, 4, 3])
y
```

Typing `?funcname` will print the documentation about the function `funcname()`.

We can tell `Python` to add two sets of numbers together. It will then add the first number from `x` to the first number from `y`, and so on. However, `x` and `y` should be the same length.
We can check their length using the `len()` function.

```python
print(len(x))
print(len(y))
x + y
```

Matrices in Python are 2D numpy arrays, they can be defined as nested list, or specifying the shape.

```python
x = np.array([[1, 2],
              [3, 4]])
x
```

The `sqrt()` function returns the square root of each element of a vector or matrix. The command
`x**2` raises each element of `x` to the power `2`; any powers are possible, including fractional or negative powers.

```python
np.sqrt(x)
```

```python
x**2
```

The `normal()` function generates a vector of random
normal variables, with the `size` argument being the sample size, and the first two arguments the mean and standard deviation.
Each time we call this function, we will get a different answer. Here we
create two correlated sets of numbers, `x` and `y`, and use
the `corrcoef()` function to compute the correlation between
them.

```python
x = np.random.normal(size=50)
y = x + np.random.normal(50, .1, size=50)
np.corrcoef(x, y)
```

By default, `normal()` creates standard normal random variables with a mean of $0$ and a standard deviation of $1$. However, the mean and standard deviation can be altered using the `loc` and `scale` arguments,
as illustrated above. Sometimes we want our code to reproduce the exact same set of random numbers; we can use the `seed()` function to do this. The `seed()` function takes an (arbitrary) integer argument.

```python
np.random.seed(1303)
np.random.normal(size=50)
```

We use `seed()` throughout the labs whenever we perform calculations involving random quantities. In general this should allow the user to reproduce our results. However,  as new versions of `numpy` become available,  small discrepancies may arise between this book and the output from `numpy`.

The `mean()` and `var()` functions can be used to compute the mean and variance of a vector of numbers. Applying `sqrt()` to the output of `var()` will give the standard deviation. Or we can simply use the
`std()` function.

```python
np.random.seed(3)
y = np.random.normal(size=100)
```

```python
np.mean(y)
```

```python
np.var(y)
```

```python
np.sqrt(np.var(y))
```

```python
np.std(y)
```

## Graphics

There exist several module to create plots and graphics. The main one is `matplotlib`.
The `plot()` function is the primary way to plot data in `matplotlib`. For instance, `plot(x, y, 'o')` produces a scatterplot of the numbers in `x` versus the numbers in
`y`. There are many additional options that can be passed in to the `plot()` function. To find out more information about the `plot()` function,
type `?plot`.

```python
import matplotlib.pyplot as plt
```

```python
x = np.random.normal(size=100)
y = np.random.normal(size=100)
plt.plot(x, y, 'o')
plt.xlabel("this is the x-axis")
plt.ylabel("this is the y-axis")
plt.title("Plot of X vs Y");
```

We will often want to save the output of a plot. We can choose among several file types using the extension of the output.

```python
plt.plot(x, y, 'o')
plt.xlabel("this is the x-axis")
plt.ylabel("this is the y-axis")
plt.title("Plot of X vs Y")
plt.savefig('Figure.pdf')
```

The function `linspace()` can be used to create a sequence of numbers. For instance, `linspace(a, b)` makes a vector of numbers between `a` and `b`.
There are many other options: for instance,  `linspace(0, 1, 10)` makes a sequence of `10` numbers that are equally spaced between `0` and `1`.

```python
x = np.linspace(1, 10, 10)
x
```

```python
x = np.linspace(-np.pi, np.pi, 50)
x
```

We  will now create some more sophisticated plots. The `contour()` function produces a contour plot in order to represent three-dimensional data; it is  like a topographical map.
 It takes three arguments:
 
 * A grid of the `x` values (the first dimension),
 * A grid of the `y` values (the second dimension), and
 * A matrix whose elements correspond
 to the `z` value (the third dimension) for each pair of (`x`, `y`) coordinates.

As with the  `plot()` function, there are many other inputs that can be used to fine-tune the output of the `contour()` function. To learn more about these, take a look at  the help file by typing `?contour`.

```python
y = x
X, Y = np.meshgrid(x, y)
Z = np.cos(Y) / (1 + X**2)
ax = plt.contour(X, Y, Z)
ax.clabel(ax.levels);
```

```python
ax = plt.contour(X, Y, Z, levels=45)
ax.clabel(ax.levels, fontsize=7);
```

```python
Z2 = (Z - Z.T)/2
ax = plt.contour(X, Y, Z2, levels=15)
ax.clabel(ax.levels);
```

The `pcolormesh()` function works the same way as `contour()`, except that it produces a color-coded plot whose colors depend on the `z` value. This is
 known as a _heatmap_, and is sometimes used to plot temperature in weather forecasts. 

```python
plt.pcolormesh(X, Y, Z2);
```

Alternatively, 3D plots can be created passing the `projection="3d"` keyword argument.
The functions `view_init` controls the angles at which the plot is viewed.

```python
fig = plt.figure(figsize=(10,10))
ax = fig.add_subplot(projection='3d')
ax.plot_wireframe(X, Y, Z2)
ax.view_init(azim=90);
```

```python
fig = plt.figure(figsize=(10,10))
ax = fig.add_subplot(projection='3d')
ax.plot_wireframe(X, Y, Z2)
ax.view_init(azim=90+30);
```

```python
fig = plt.figure(figsize=(10,10))
ax = fig.add_subplot(projection='3d')
ax.plot_wireframe(X, Y, Z2)
ax.view_init(azim=90+30, elev=20);
```

```python
fig = plt.figure(figsize=(10,10))
ax = fig.add_subplot(projection='3d')
ax.plot_wireframe(X, Y, Z2)
ax.view_init(azim=90+30, elev=70);
```

```python
fig = plt.figure(figsize=(10,10))
ax = fig.add_subplot(projection='3d')
ax.plot_wireframe(X, Y, Z2)
ax.view_init(azim=90+30, elev=40);
```

## Indexing Data

We often wish to examine part of a set of data. Suppose that our data is stored in the matrix `A`.

```python
A = np.array(range(1, 17)).reshape((4, 4))
A
```

Then, typing

```python
A[1, 2]
```

will select the element corresponding to the second row and the third column. The first number after the open-bracket symbol `[`
 always refers to the row, and the second number always refers to the column. We can also select multiple rows and columns at a time using slices and fancy indexing

```python
A[np.ix_([0,2], [1,3])]
```

```python
A[0:3, 1:4]
```

```python
A[0:2, :]
```

```python
A[:, 0:2]
```

The last two examples include either `:` for the columns or `:` for the rows. These  indicate that `numpy` should include all columns or all rows, respectively. `numpy` treats a single row or column of a matrix as a vector.

```python
A[0, :]
```

The `shape` attribute outputs the number of rows followed by the number of columns of a given matrix.

```python
A.shape
```

## Loading Data

For most analyses, the first step involves importing a data set into `Python`. 
For manipulating tabular datasets, we will be using the `pandas`library.
The `read_csv()` function is one of the primary ways to do this. The documentation contains details about how to use this function.
We can use the function `write_csv()` to export data.

Before attempting to load a data set, we must make sure that `Python`
knows to search for the data in the proper directory. 

We begin by loading in the `Auto` data set. To illustrate the `read_csv()` function, we load it now from a text file, `Auto.csv`, which you can find on the textbook website.
The following command will load the `Auto.csv` file into `Python` and store it as an object called `Auto`,  in a format referred to as a _data frame_. The `head()` function can also be used to view the first few rows of the data.

```python
import pandas as pd
```

```python
Auto = pd.read_table('data/Auto.data', delim_whitespace=True)
Auto.info()
```

```python
Auto.head()
```

Note that Auto.data is simply a text file, which you could
alternatively open on your computer using a standard text editor. It
is often a good idea to view a data set using a text editor or other
software such as Excel before loading it into `Python`.

The data set also includes a number of missing observations, indicated by
a question mark `?`. Missing values are a common occurrence in real data sets. Using the option `na.strings` tells `pandas` that any time it sees a particular character or set of characters (such as a question mark), it should
be treated as a missing element of the data matrix. 

```python
Auto = pd.read_table("data/Auto.data", na_values='?', delim_whitespace=True)
Auto.info()
```

```python
Auto.shape
```

The `shape` attribute tells us that the data has $397$ observations, or rows, and nine variables, or columns. There are various ways to deal with the missing data.  In this case,   only five of the rows contain missing observations,
and  so we choose to use the `dropna()` function to simply remove these rows.

```python
Auto = Auto.dropna()
Auto.shape
```

Once the data are loaded correctly, we can use `columns` to check the variable names.

```python
Auto.columns
```

## Additional Graphical and Numerical Summaries

We can use the `plot()` function to produce *scatterplots*  of the quantitative variables.

```python
Auto.plot.scatter('cylinders', 'mpg')
```

```python
Auto.boxplot(by='cylinders', column='mpg');
```

The `hist()` function can be used to plot a _histogram_.

```python
Auto.mpg.hist(bins=15, color='r');
```

The `scatter_matrix()` function creates a *scatterplot matrix*, i.e. a scatterplot for every pair of variables. We can also produce scatterplots for just a subset of the variables.

```python
pd.plotting.scatter_matrix(Auto, figsize=(15, 15));
```

```python
pd.plotting.scatter_matrix(Auto[['mpg', 'displacement', 'horsepower', 'weight', 'acceleration']], figsize=(15, 15));
```

The `describe()` function produces a numerical summary of each variable in a particular data set.

```python
Auto.describe(include='all')
```

For qualitative variables such as `name`,  `pandas` will list the number of observations and the most common category. We can also produce a summary of just a single variable.

```python
Auto.name.describe()
```
